```python
#mini dictionary daraus bauen
import time
leereliste = []

for beben in erdbeben:
    place = beben['properties']['place']
    magnitude = beben['properties']['mag']
    zeit = beben['properties']['time']
    mini_dict = {'place':place,
                'magnitude': magnitude,
                'time': time.ctime(int(zeit))} #unix-time umgerechnet, !, variable deutsch(zeit), not time!
    print(mini_dict)
    leereliste.append(mini_dict)
    
pd.DataFrame(leereliste)
```
